```python
# Create a function to define those who are mid-life (30-45)
def male_female_midlife(credit_holder):
    age, sex = credit_holder
    
    if age > 40:
        return 'midlifeplus'
    else:
        return sex
```

```python
credit['person'] = credit[['age', 'sex']].apply(male_female_midlife, axis=1)
```

```python
fig = sns.FacetGrid(credit, hue='sex', aspect=4)
fig.map(sns.kdeplot, 'age', shade=True)
oldest = credit['age'].max()
fig.set(xlim=(0,oldest))
fig.set(title='Distribution of Age Grouped by Gender')
fig.add_legend()
```

```python
fig = sns.FacetGrid(credit, hue='person', aspect=4)
fig.map(sns.kdeplot, 'education', shade=True)
oldest = credit['education'].max()
fig.set(xlim=(0,oldest))
fig.add_legend()
```
